```r
#loading the libraries
library(timeSeries)
library(readxl)
library(ggplot2)
```

```r
#importing the data and skiping the irrelevant rows
data <- read_excel("/home/mngadil/Downloads/Natasha_Unemp.xlsx")
```

```r
# I have converted the data types from char type for convenient time series working 
data$Date <- as.Date(data$Date, format = "%Y")
data$Rate <- as.numeric(data$Rate)
```

```r
# This is unnecessary but I have changed column of y-coordinate, you can just uncomment these two lines below for that
# colnames(data)[colnames(data) == "KBP7019L"] <- "Rate"
# colnames(data)
```

```r
df <- data.frame(data)
# data
```

```r

ggplot(df, aes(x = Date, y = Rate)) +
geom_point(size=5) +  
geom_line(size=2, color="blue") +
  labs(x = "Date", y = "Unemployment") +
  ggtitle("Seasonal Time Series Data") + theme_bw()
```
